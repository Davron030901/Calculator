```python
# prompt: vector norm

import numpy as np

def vector_norm(vector, order=2):
  """
  Calculates the vector norm of a given vector.

  Args:
    vector: A NumPy array representing the vector.
    order: The order of the norm (default is 2 for Euclidean norm).

  Returns:
    The vector norm as a float.
  """
  return np.linalg.norm(vector, ord=order)

# Example usage
vector_a = np.array([1, 2, 3])
norm_a = vector_norm(vector_a)
print(f"The Euclidean norm of vector_a is: {norm_a}")

norm_a_1 = vector_norm(vector_a, order=1)
print(f"The L1 norm of vector_a is: {norm_a_1}")

norm_a_inf = vector_norm(vector_a, order=np.inf)
print(f"The L-infinity norm of vector_a is: {norm_a_inf}")
```
