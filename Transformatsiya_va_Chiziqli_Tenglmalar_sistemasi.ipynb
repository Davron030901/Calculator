```python
def transform(A,b,x):
  m,n=A.shape
  A_aug=np.vstack([A,np.zeros(n)])
  b_aug,x_aug=np.vstack([b,[1]]),np.vstack([x,[1]])
  T=np.hstack([A_aug,b_aug])
  y_aug=T@x_aug
  y=y_aug[:-1]
  return y
```

```python
A=np.array([[1,0.1],
            [1,1]])
b=np.array([[0],[0]])
x=np.array([[1],[1.5]])
y=transform(A,b,x)
print(y)
```

##Siljitish matritsasi

```python
A_shear=np.array([[1,1.5],
            [0,1]])
```

##Aylantish matritsasi (45 gradus)

```python
theta=np.pi/4
A_rotation=np.array([[np.cos(theta),-np.sin(theta)],
            [np.sin(theta),np.cos(theta)]])
```

##Aks matritsasi (y-o'qi bo'yicha

```python
A_reflection=np.array([[-1,0],
            [0,1]])
```

##Masshtablash matritsasi

```python
A_scaling=np.array([[2,0],
            [0,0.5]])
```

```python
transformations=[
    ("Shear",A_shear),
    ("Rotation",A_rotation),
    ("Reflection",A_reflection),
    ("Scaling",A_scaling)
]
```

```python
x = np.array([[1], [1]])
b = np.array([[0], [0]])



fig, axs = plt.subplots(1, 4, figsize=(18, 3))



for ax, (title, A) in zip(axs.flatten(), transformations):
    y = transform(A, b, x)

    ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')
    ax.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed Vector')

    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(True)

    ax.set_title(title)
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.legend()

plt.tight_layout()
plt.show()
```
